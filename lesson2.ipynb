1. Matrisa näme we ol näme üçin gerek?
Matrisa — bu sanlaryň tablisasydyr. Neýron torlarynda:

Input (X): Giriş maglumatlarynyň matrisasy.

Weights (W): Ähli neýronlaryň agramlarynyň matrisasy.

Eger biziň 3 sany girişimiz we 1 neýronymyz bar bolsa, hasaplama şeýle bolar:

```python
import numpy as np

# 1. Giriş maglumatlary (meselem: howanyň temperaturasy, çyglylygy we basyşy)
# Bularyň hemmesi bilelikde bir 'Vector' (1-ölçegli matrisa)
inputs = np.array([25, 60, 1013]) 

# 2. Agramlar (Her giriş üçin aýratyn agram)
weights = np.array([0.5, -0.2, 0.1])

# 3. Bias (Süýşme/Gidim) - Bu neýrona has çeýe bolmaga kömek edýän goşmaça san
bias = 2

# 4. Hasaplama (Dot Product)
# output = (25 * 0.5) + (60 * -0.2) + (1013 * 0.1) + 2
output = np.dot(inputs, weights) + bias

print(f"Neýronyň çykyşy: {output}")
```

"Bias" näme?
Ýokardaky kodda biz bias goşduk. Bu näme üçin gerek?Matematiki taýdan $y = wx$ diýsek, grafik elmydama koordinata başlangyjyndan ($0,0$) geçmeli bolýar. Ýöne biziň maglumatlarymyz elmydama noldan başlamaýar. bias (b) goşmak bilen biz grafigi ýokary ýa-da aşak süýşürip bilýäris: $y = wx + b$. Bu neýrona maglumatlara has gowy "uýgunlaşmaga" mümkinçilik berýär.

```python
import numpy as np

# 4 sany giriş (Inputs)
X = np.array([1.0, 2.0, 3.0, 2.5])

# 3 sany neýron üçin agramlar (Weights matrisasy: 3x4 ölçegli)
# Her neýron 4 girişli bolansoň, her setirde 4 agram bar.
W = np.array([
    [0.2, 0.8, -0.5, 1.0],  # 1-nji neýronyň agramlary
    [0.5, -0.91, 0.26, -0.5], # 2-nji neýronyň agramlary
    [-0.26, -0.27, 0.17, 0.87] # 3-nji neýronyň agramlary
])

# Her neýron üçin 1 sany Bias
biases = np.array([2.0, 3.0, 0.5])

# Gatlagyň hasaplamasy
layer_outputs = np.dot(W, X) + biases

print("Gatlagyň çykyşy (3 neýronyň jogaby):")
print(layer_outputs)
```

Bu sapagyň iň möhüm ýeri:
np.dot(W, X): Bu funksiýa ähli köpeltmek we goşmak amallaryny paralel ýerine ýetirýär.

Ölçegler (Shapes): Matrisalar bilen işlenende iň kyn ýeri ölçegleri dogry sazlamakdyr. Eger giriş 4 sany bolsa, agramlar hem 4 sany bolmaly.
